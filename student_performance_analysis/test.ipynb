# 1. Préparation des données

```python
import pandas as pd

data = {'solde': [2000, 50000, 1500, 45000, 2200, 38000],
'age': [20, 45, 22, 50, 25, 40],
'achat': [0, 1, 0, 1, 0, 1]}
df = pd.DataFrame(data)
df.head()
```

# Diagramme de dispersion (scatter plot)

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df, 
    x='age', 
    y='solde', 
    hue='achat',
    style='achat',
    s=200,          
    palette='coolwarm'
)

plt.title("Relation entre l'âge, le solde et l'achat", fontsize=15)
plt.xlabel("Âge du client")
plt.ylabel("Solde sur le compte")
plt.show()
```

# 2. Séparation Features (X) et Cible (y)

```python
X = df[['solde', 'age']]    
y = df['achat']
X.head()
```

# 3. Découpage Train/Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.head()
```

# 4. Mise à l'échelle (Scaling)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(X_train_scaled)
```

# 5. Création et entraînement du modèle

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

 # 6. Prédiction

```python
prediction = model.predict(scaler.transform([[3000, 30]]))
print(f"Prédiction d'achat : {'Oui' if prediction[0] == 1 else 'Non'}")
```
