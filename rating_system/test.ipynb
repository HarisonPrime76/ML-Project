# Initialisation de notre base de données (dictionnaire)

```python

produit = {
    "nom": "Ordinateur Portable",
    "total_points": 10,
    "nombre_votes": 5
}

print(f"Produit : {produit['nom']}")
print(f"Moyenne actuelle : {produit['total_points'] / produit['nombre_votes']}/5")
```

# Fonction de traitement logique

```python
def ajouter_evaluation(article, note):
    if not (1 <= note <= 5):
        return "Erreur : La note doit être comprise entre 1 et 5."
    
    article["total_points"] += note
    article["nombre_votes"] += 1
    
    # Calcul de la nouvelle moyenne
    moyenne = article["total_points"] / article["nombre_votes"]
    return round(moyenne, 2)
```

# Simulation d'un vote

```python
nouvelle_note = 3.5
nouvelle_moyenne = ajouter_evaluation(produit, nouvelle_note)

print(f"Merci pour votre vote de {nouvelle_note} !")
print(f"La nouvelle note moyenne est de : {nouvelle_moyenne}/5")
```

# Affichage visuel 

```python
def afficher_etoiles(moyenne):
    nb_etoiles_pleines = int(moyenne)
    return "★" * nb_etoiles_pleines + "☆" * (5 - nb_etoiles_pleines)

moyenne_actuelle = produit["total_points"] / produit["nombre_votes"]
print(f"Note visuelle : {afficher_etoiles(moyenne_actuelle)}")
```
